```python
df = df.dropna()
df.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.countplot(y=df["country"], order=df["country"].value_counts().index)
plt.title("Country-wise Record Distribution")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

```python
top_10_countries = df["country"].value_counts().nlargest(10).index

filtered_df = df[df["country"].isin(top_10_countries)]

# Plot
plt.figure(figsize=(12, 6))
sns.countplot(y=filtered_df["country"], order=top_10_countries)
plt.title("Top 10 Countries by Record Count")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

```python
below_10_countries = df["country"].value_counts().nsmallest(10).index

filtered_df = df[df["country"].isin(below_10_countries)]

# Plot
plt.figure(figsize=(12, 6))
sns.countplot(y=filtered_df["country"], order=below_10_countries)
plt.title("Below 10 Countries by Record Count")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df["polygon_level"].value_counts().plot.pie(autopct="%1.1f%%", startangle=140, cmap="Pastel1")
plt.title("Proportion of Polygon Levels")
plt.ylabel("")
plt.show()
```

```python
counts = df['home_to_ping_distance_category'].value_counts()


plt.figure(figsize=(10, 6))
plt.barh(counts.index, counts.values, color=plt.cm.Paired.colors[:len(counts)])


plt.xlabel('Count')
plt.ylabel('Distance Category')
plt.title("Distribution of Distance Categories")

for index, value in enumerate(counts.values):
    plt.text(value + 1, index, str(value), va='center', fontsize=12)

plt.gca().invert_yaxis()  
plt.show()
```

```python
df['ds'] = pd.to_datetime(df['ds'])
daily_ping_fraction = df.groupby("ds")["distance_category_ping_fraction"].sum()
daily_ping_fraction.plot(title="Daily Ping Fraction Trend")
plt.show()
```

```python
print(sorted(df['country'].unique()))
```

```python

negative_values = df[df["distance_category_ping_fraction"] < 0]
print(f"\nNumber of negative values in distance_category_ping_fraction: {len(negative_values)}")
```

```python
country_data = df.groupby('country')['distance_category_ping_fraction'].mean(numeric_only=True).sort_values()

# Plot
plt.figure(figsize=(12, 6))
plt.bar(country_data.index, country_data.values, color='skyblue')
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Avg Ping Fraction")
plt.title("Country-Level Average Ping Fraction")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import numpy as np

top_10_countries = df.groupby('country')['distance_category_ping_fraction'].mean(numeric_only=True).nlargest(10)
colors = plt.cm.Paired(np.linspace(0, 1, len(top_10_countries)))

# Plot
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries.index, top_10_countries.values, color=colors)
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Avg Ping Fraction")
plt.title("Country-Level Average Ping Fraction")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
bottom_10_countries = df.groupby('country')['distance_category_ping_fraction'].mean(numeric_only=True).nsmallest(10)

colors = plt.cm.Paired(np.linspace(0, 1, len(top_10_countries)))

plt.figure(figsize=(12, 6))
plt.bar(bottom_10_countries.index, bottom_10_countries.values, color=colors)
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Avg Ping Fraction")
plt.title("Country-Level Average Ping Fraction")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
india_data = df[df['country'] == 'IND']['distance_category_ping_fraction'].dropna()

plt.figure(figsize=(8, 6))
plt.boxplot(india_data, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Distribution of Ping Fractions in India")
plt.ylabel("Ping Fraction")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
